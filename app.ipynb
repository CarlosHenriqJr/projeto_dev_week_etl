```python
# Importações necessárias
import pandas as pd
import re
```

```python
# Carregar os dados
copa_data = pd.read_excel("database.xlsx")
```

```python
# Substituir '5+' por 6 na coluna TimeA_Gols
copa_data['TimeA_Gols'] = copa_data['TimeA_Gols'].replace('5+', 6)

# Converter a coluna TimeB_Gols para o tipo de dados inteiro
copa_data['TimeA_Gols'] = copa_data['TimeA_Gols'].astype(int)

# Verificar os valores únicos da coluna TimeB_Gols após a correção
valores_unicos_timeb_gols = copa_data['TimeA_Gols'].unique()
```

```python
# Substituir '5+' por 6 na coluna TimeB_Gols
copa_data['TimeB_Gols'] = copa_data['TimeB_Gols'].replace('5+', 6)

# Converter a coluna TimeB_Gols para o tipo de dados inteiro
copa_data['TimeB_Gols'] = copa_data['TimeB_Gols'].astype(int)

# Verificar os valores únicos da coluna TimeB_Gols após a correção
valores_unicos_timeb_gols = copa_data['TimeB_Gols'].unique()
```

```python
# Defina uma função para criar a nova coluna "Mais 2.5"
def criar_coluna_mais_2_5(row):
    if row['TimeA_Gols'] + row['TimeB_Gols'] > 2.5:
        return "SIM TEVE MAIS DE 2.5 Gols"
    else:
        return "NÃO TEVE MAIS DE 2.5 Gols"

# Aplique a função para criar a nova coluna
copa_data['Mais_2.5'] = copa_data.apply(criar_coluna_mais_2_5, axis=1)
```

```python
# Defina uma função para criar a nova coluna "Mais 2.5"
def criar_coluna_mais_3_5(row):
    if row['TimeA_Gols'] + row['TimeB_Gols'] > 3.5:
        return "SIM TEVE MAIS DE 3.5 Gols"
    else:
        return "NÃO TEVE MAIS DE 3.5 Gols"

# Aplique a função para criar a nova coluna
copa_data['Mais_3.5'] = copa_data.apply(criar_coluna_mais_3_5, axis=1)
```

```python
# Defina uma função para criar a nova coluna "Mais 2.5"
def criar_coluna_Ambos_Marcam(row):
    if (row['TimeA_Gols'] > 0) & (row['TimeB_Gols'] > 0):
        return "SIM, AMBOS MARCARAM"
    else:
        return "NÃO, AMBOS NÃO MARCARAM"

# Aplique a função para criar a nova coluna
copa_data['Ambos_Marcam'] = copa_data.apply(criar_coluna_Ambos_Marcam, axis=1)
```

```python
# Selecionar as colunas relevantes para visualização
colunas_relevantes = ['TimeA', 'TimeB', 'TimeA_Gols', 'TimeB_Gols','Mais_2.5','Mais_3.5','Ambos_Marcam']

copa_data = copa_data[colunas_relevantes]

# Mostrar os primeiros registros
copa_data.head(10)
```
